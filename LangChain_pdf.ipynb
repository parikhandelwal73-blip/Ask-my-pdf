Load the PDF file and extract its contents.

```python
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader('PDF_Chatbot_RAG_Learning_Roadmap.pdf')
docs=loader.load()
print(docs)
print(len(docs))
```

Split the loaded PDF text into smaller overlapping chunks.

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
splits=text_splitter.split_documents(docs)
print(splits[0])
print(splits[1])
print(len(splits))
```

Initialize the Google Gemini embedding model to convert text chunks into vector embeddings.

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
embedding = GoogleGenerativeAIEmbeddings(model="gemini-embedding-2-preview")
```

Create a Chroma vector database and store the vector embeddings of the text chunks.

```python
from langchain_chroma import Chroma
vectorstore=Chroma.from_documents(documents=splits,embedding=embedding)
```

Create a retriever from the Chroma vector database to fetch relevant document chunks.

```python
retriever=vectorstore.as_retriever()
```

Define the prompt template for the RAG-based AI assistant.

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template("""
You are an expert AI assistant.

First, use the provided context to answer the question.

If the context contains the answer, answer from it.

If the context is incomplete or does not contain the answer, use your own knowledge to provide a helpful response.

If you use your own knowledge, clearly mention:

"Note: The following information is based on my general knowledge and not on the provided PDF."

Context:
{context}

Question:
{question}

Instructions:
1. Give a clear, detailed, and easy-to-understand answer.
2. If the answer comes from the PDF, answer using the PDF content.
3. If the PDF does not contain enough information, use your own knowledge and include the note above.
4. After answering, suggest 3 to 5 closely related topics that the user may also find useful.
5. End your response with:

Related Topics:
- Topic 1
- Topic 2
- Topic 3
- Topic 4
- Topic 5

Would you like to learn more about any of these topics?

Type your response below:
Yes / No

If the user replies "Yes", continue explaining the selected topic.
If the user replies "No", respond with:
"Thank you for using this AI assistant. I hope I was able to help you. Have a wonderful day!"

Answer:
""")
```

Initialize the Gemini language model to process prompts and generate AI responses.

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    temperature=0
)
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

Combine the retrieved document chunks into a single text string separated by blank lines.

```python
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)
```

Build the complete RAG pipeline to retrieve relevant documents, generate a response using Gemini, and return the answer as plain text.

```python
rag_chain=({"context":retriever| format_docs,"question":RunnablePassthrough()}|prompt|llm|StrOutputParser())
```

Process user questions, generate AI responses, and support follow-up interactions.

```python
import re
def valid_question(question):
    question = question.strip()
    # Empty input
    if question == "":
        return False
    # Greetings
    greetings = {
        "hi",
        "hello",
        "hey",
        "good morning",
        "good afternoon",
        "good evening"
    }
    if question.lower() in greetings:
        return False
    # Single character
    if len(question) == 1:
        return False
    # Numbers only
    if question.isdigit():
        return False
    # Symbols only
    if re.fullmatch(r"[^\w]+", question):
        return False
    return True
while True:
    question = input("\nAsk your question (or type 'exit' to quit): ").strip()
    # Exit chatbot
    if question.lower() in ["exit", "quit"]:
        print("\nThank you for using this AI assistant. Have a wonderful day!")
        break
    # Validate question
    if not valid_question(question):
        print("\nSorry! Invalid question. Please enter a complete and meaningful question.")
        continue
    # Get answer from RAG
    response = rag_chain.invoke(question)
    print("\nAnswer:\n")
    print(response)
    # Ask if the user wants to learn more
    while True:
        choice = input("\nWould you like to know more? (Yes/No): ").strip().lower()
        if choice in ["yes", "y"]:
            topic = input("\nEnter the related topic you want to learn about: ").strip()
            if not valid_question(topic):
                print("\nPlease enter a valid topic.")
                continue
            response = rag_chain.invoke(f"Explain in detail about {topic}")
            print("\nAnswer:\n")
            print(response)
        elif choice in ["no", "n"]:
            print("\nOkay! Ask me another question.")
            break
        else:
            print("\nPlease type only Yes/Y or No/N.")
```
